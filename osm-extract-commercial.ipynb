```python
import xml.etree.ElementTree as ET
import pandas as pd
```

## Make supermarket

```python
mytree = ET.parse('map.xml')
myroot = mytree.getroot()

supermarkets = []
for i, node in enumerate(myroot.findall('node')):
    is_supermarket = False
    
    kv = {tag.attrib['k']: tag.attrib['v'] for tag in node.findall('tag')}
    if kv.get('shop') == 'supermarket':
        item = {
            'lat': node.attrib['lat'],
            'lon': node.attrib['lon'],
            'brand': kv.get('brand'),
            'branch': kv.get('branch'),
            'wheelchair': kv.get('wheelchair'),
            'opening_hours': kv.get('opening_hours')
        }
        supermarkets.append(item)
df = pd.DataFrame(supermarkets).dropna(subset=['brand', 'opening_hours'])
df.to_csv('supermarkeder_kbh.csv', index=False)
df
```

## Make commercial

```python
mytree = ET.parse('map.xml')
myroot = mytree.getroot()

commercial_kbh = []
for i, node in enumerate(myroot.findall('node')):
    is_supermarket = False
    
    kv = {tag.attrib['k']: tag.attrib['v'] for tag in node.findall('tag')}
    kind = kv.get('shop') or kv.get('amenity') or kv.get('tourism')
    if kind:
        category = [x for x in kv.keys() if x in ['shop', 'amenity', 'tourism']][0]
        item = {
            'lat': node.attrib['lat'],
            'lon': node.attrib['lon'],
            'name': kv.get('name'),
            'kind': kind,
            'category': category
            
        }
        commercial_kbh.append(item)
df2 = pd.DataFrame(commercial_kbh)
df2.to_csv('commercial_kbh.csv', index=False)
df2.head()
```
